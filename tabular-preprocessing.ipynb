```python
# -----------------------------
# 1. Define columns to KEEP
# -----------------------------
important_columns = [
    "Patient Id",
    "Gender",
    "Age",   
    "height  (meter)",
    "Weight (KG) ",
    "BMI: ",
    "Smoker",
    "Alcoholic",
    "Diabetic",
    "Hypothyroidism",
    "Number of Pregnancies",
    "Estrogen Use",
    "History of Fracture",
    "Family History of Osteoporosis",
    "Obesity",
    "Diagnosis"
]

# -----------------------------
# 2. Filter the dataframe
# -----------------------------
df_clean = df[important_columns]

# -----------------------------
# 3. Verify result
# -----------------------------
df_clean.info()
```

```python
for col in df_clean.columns:
    print(f"\n--- {col} ---")
    print(df_clean[col].unique())
```

```python
# ----------------------------------------
# 1. Fill NULL values in Number of Pregnancies with 0
# ----------------------------------------
df_clean["Number of Pregnancies"] = df_clean["Number of Pregnancies"].fillna(0)

# ----------------------------------------
# 2. Binary Encode All Yes/No Columns
# ----------------------------------------
yes_no_columns = [
    "Smoker",
    "Alcoholic",
    "Diabetic",
    "Hypothyroidism",
    "Estrogen Use",
    "Family History of Osteoporosis"
]

for col in yes_no_columns:
    df_clean[col] = df_clean[col].map({"yes": 1, "no": 0})

# ----------------------------------------
# 3. Convert History of Fracture to Only Yes/No → Then Encode
# ----------------------------------------

# First convert everything that is NOT 'no' into 'yes'
df_clean["History of Fracture"] = df_clean["History of Fracture"].astype(str).str.lower()

df_clean["History of Fracture"] = df_clean["History of Fracture"].apply(
    lambda x: 0 if x == "no" else 1
)

# ----------------------------------------
# 4. Verify Output
# ----------------------------------------
print(df_clean[[
    "Smoker", "Alcoholic", "Diabetic", "Hypothyroidism",
    "Estrogen Use", "Family History of Osteoporosis",
    "History of Fracture", "Number of Pregnancies"
]].head())

print("\nMissing values after preprocessing:")
print(df_clean.isnull().sum())
```
